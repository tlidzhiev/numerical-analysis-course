$$

$$

$$

$$
Вычислите ошибку решения системы линейных уравнений:
$$

$\phi_1(x), ... , \phi_n(x)$ - набор базисных функций.

Искомая функция ищется в виде линейной комбинации:
$$
f(x) \approx \sum_{k=1}^{n} c_k \phi_k(x)
$$
Коэффициенты $c_k$ можно искать, минимизируя среднеквадратичную ошибку в известных точках.
Тогда задача сводится к системе линейных уравнений:
$$
Ac = b
$$
Элементы матрицы $A$ вычисляются следующим образом:
$$
A_{ij} = \sum_{\alpha=1}^{N}\frac{\phi_i(x_\alpha) \phi_j(x_\alpha)}{N}
$$
здесь $N$ - число точек, в которых известны значения функции.
$$
b_{i} = \sum_{\alpha=1}^{N}\frac{\phi_i(x_\alpha) y_\alpha}{N}
$$
